```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def token(text):
    token = nltk.word_tokenize(text)
    lis = [stemmer.stem(w) for w in token]
    return " ".join(lis) 
```

```python
df["text"].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(analyzer="word",stop_words="english")
```

```python
matrix = tfidf.fit_transform(df["text"])
```

```python
matrix.shape
```

```python
similarity = cosine_similarity(matrix)
```

```python
df["song"].tail(5)
```

```python
df[df["song"] == "Audrey"].index[0]
```

```python
# sorted(list(enumerate(similarity[19999])), reverse=True, key=lambda x: x[1])
```

```python
def recommender(song_name):
    idx = df[df["song"] == song_name].index[0]
    distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])
    song = []
    for song_id in distance[1:5]:
        song.append(df.iloc[song_id[0]].song)
    return song
```

```python
recommender("Audrey")
```
